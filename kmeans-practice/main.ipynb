```python
#importlar
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import word_tokenize

import csv

from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer

#nltk.download('stopwords')
#nltk.download('punkt')
```

```python
def dataPreprocess():
    r = csv.reader(open('BBC News Train.csv')) # Here your csv file
    lines = list(r)
    
    stop_words = set(stopwords.words('english')) 
    
    ps = PorterStemmer() 
    
    j = 0
    for article2 in lines:
        article = article2[1]
        words = word_tokenize(article)

        words=[word.lower() for word in words if word.isalpha()]
        words = [ps.stem(word) for word in words if word.isalpha()]
        wordsFiltered = []

        for w in words: 
            if w not in stop_words:
                wordsFiltered.append(w)
        words = wordsFiltered[:]                      

        lines[j][1] = words[:]
        j += 1
        print("adım:", j)

    for article in lines:
        tmpStr = ""
        for word in article[1]:
            tmpStr += word + " "
        article[1] = tmpStr

    writer = csv.writer(open('BBC News Train Preprocessed.csv', 'w'), lineterminator='\n')
    writer.writerows(lines)
    
#dataPreprocess()
```

```python
def findLowFrequencyWords():
    read = pd.read_csv('BBC News Train Preprocessed.csv', sep=r'\s*,\s*',
                 header=0, encoding='ascii', engine='python')
    texts = np.array(read['text'])

    cv = CountVectorizer() 
    word_count_vector = cv.fit_transform(texts)

    indices = cv.vocabulary_
    #(1490, 15718)
    wordCounts = word_count_vector.toarray()

    #amount = 0
    lowFrequencyWords = []
    for name in cv.get_feature_names():
        count = 0
        for i in range(1490):
            count += wordCounts[i][indices[name]]
        if (count <= 3):
            #print(name, count)
            #amount += 1
            lowFrequencyWords.append(name)
    #print(lowFrequencyWords)
    #8833
    #print(amount)
    return lowFrequencyWords

lowFrequencyWords = findLowFrequencyWords()
print(lowFrequencyWords)
```

```python
def eliminateLowFrequencyWords(lowFrequencyWords):
    r = csv.reader(open('BBC News Train Preprocessed.csv')) # Here your csv file
    lines = list(r)
    
    j = 0
    for article2 in lines:
        article = article2[1]
        words = word_tokenize(article)

        wordsFiltered = []
        for w in words: 
            if w not in lowFrequencyWords:
                wordsFiltered.append(w)
        words = wordsFiltered[:]                      

        lines[j][1] = words[:]
        j += 1
        print("adım:", j)

    for article in lines:
        tmpStr = ""
        for word in article[1]:
            tmpStr += word + " "
        article[1] = tmpStr

    writer = csv.writer(open('BBC News Train ExtraPreprocessed.csv', 'w'), lineterminator='\n')
    writer.writerows(lines)

eliminateLowFrequencyWords(lowFrequencyWords)
print("bitti")
```

```python
def kMeans(X):
    km = KMeans(
        n_clusters = 5, init = 'k-means++', #hedef küme: 5, centerları akıllı bir şekilde seç
        n_init = 10, max_iter = 100 #10 kere birbirinden bağımsız olarak çalıştır en düşük hatalıyı al?, en fazla 100 iterasyon
    )
    Y = km.fit_predict(X)
    
    return Y
```

```python
#tfIdf çıkarma

#sadece text kısmını al
read = pd.read_csv('BBC News Train ExtraPreprocessed.csv', sep=r'\s*,\s*',
                 header=0, encoding='ascii', engine='python')
texts = np.array(read['text'])

#feature vector oluştur
vectorizer = TfidfVectorizer(use_idf=True) 
vectors = vectorizer.fit_transform(texts)

print("done")

#df = pd.DataFrame(vectors[0].T.todense(), index=tfidf_vectorizer.get_feature_names(), columns=["tfidf"]) 
#df.sort_values(by=[""],ascending=False)
```

```python
#Kümeleme ve başarı hesabı
r = csv.reader(open('BBC News Train ExtraPreprocessed.csv')) # Here your csv file
lines = list(r)

def clusterAndMatchLabels(vectors):
    Y = kMeans(vectors)

    clusters = [[], [], [], [], []]
    for i in range(len(Y)):
        arr = [i, lines[i+1][2]]
        clusters[Y[i]].append(arr)
    return clusters
    
def countClusterTopics(clusters):
    result = []
    for i in range(5):
        arr = [0, 0, 0, 0, 0]
        for j in range(len(clusters[i])):
            if (clusters[i][j][1] == 'politics'):
                arr[0] += 1
            elif (clusters[i][j][1] == 'entertainment'):
                arr[1] += 1
            elif (clusters[i][j][1] == 'tech'):
                arr[2] += 1
            elif (clusters[i][j][1] == 'sport'):
                arr[3] += 1
            elif (clusters[i][j][1] == 'business'):
                arr[4] += 1
        result.append(arr)
    return result

def calculateAccuracy(result):
    politics = 0    
    entertainment = 0
    tech = 0
    sport = 0
    business = 0
    correctValues = [0, 0, 0, 0, 0]

    total = 0
    correct = 0

    #print("#Kümelerin tek bir tipi kümeleme başarısı#")
    for i in range(5):
        amount = 0
        print("\nCluster %d:" % i)
        print("politics:", result[i][0])
        politics += result[i][0]
        print("entertainment:", result[i][1])
        entertainment += result[i][1]
        print("tech:", result[i][2])
        tech += result[i][2]
        print("sport:", result[i][3])
        sport += result[i][3]
        print("business:", result[i][4])
        business += result[i][4]

        for j in range(5):
            if (result[i][j] == max(result[i])):
                correctValues[j] += result[i][j]
                break

        amount = result[i][0] + result[i][1] + result[i][2] + result[i][3] + result[i][4]
        total += amount
        correct += max(result[i])
        print("total:", amount, "correct:", max(result[i]), "wrong:", amount - max(result[i]))
        #print("accuracy:", max(result[i]) / amount)

    #print("\n----------")
    print("total:", total, "correct:", correct, "wrong:", total - correct)
    #print("accuracy:", correct / total)
    #print("----------")

    print("\n#Sınıfların tek bir clusterda toplanma başarısı#")
    print("\npolitics:", correctValues[0] / politics, "\nentertainment:", correctValues[1] / entertainment,
          "\ntech:", correctValues[2] / tech, "\nsport:", correctValues[3] / sport,
          "\nbusiness:", correctValues[4] / business)
    print("----------")
    acc = [(correctValues[0] / politics + correctValues[1] / entertainment +
                    correctValues[2] / tech + correctValues[3] / sport +
                    correctValues[4] / business) / 5, 
            correctValues[0] / politics,
            correctValues[1] / entertainment,
            correctValues[2] / tech,
            correctValues[3] / sport,
            correctValues[4] / business
           ]
    print("total:", acc[0])
    print("----------")
    return acc

acc = [0, 0, 0, 0, 0, 0]
amount = 10
for i in range(amount):
    print("\n", i+1,". KMeans:")
    newAcc = calculateAccuracy(
        countClusterTopics(
            clusterAndMatchLabels(vectors) 
        ) 
    )
    acc[0] += newAcc[0]
    acc[1] += newAcc[1]
    acc[2] += newAcc[2]
    acc[3] += newAcc[3]
    acc[4] += newAcc[4]
    acc[5] += newAcc[5]

print("\n\n#Sınıfların tek bir clusterda toplanma başarısı (Ortalama)#")
print("politics:", acc[1] / amount, "\nentertainment:", acc[2] / amount,
      "\ntech:", acc[3] / amount, "\nsport:", acc[4] / amount,
      "\nbusiness:", acc[5] / amount)
print("----------")
print("final:", acc[0] / amount)
```
